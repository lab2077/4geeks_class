```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
from scipy.stats import binom,poisson,norm
```

```python
#En un lanzamiento de 4 modenas, cual es la probabilidad de obtener 2 caras P(X=2)
#Cual es la probabilidad de obtener 3 o menos caras P(X<=3) esto es = P(X=0)+P(X=1)+P(X=2)+P(X=3) .
#espacio_muestra = 0,1,2,3,4
n=4
p=0.5
results=[0,1,2,3,4] #list(range(n+1)) # list me permite hacer un lista y range me permite establecer el rango
freqs=[binom.pmf(i,n,p) for i in results] #i n p - exitos eventos probabi;idad, PMF es para valores discretos(contables)
freqs

#Grafico
plt.bar(results,freqs)
for i, val in enumerate(freqs): #enumerate nos devuelve como primer termino el indice de la coleccion, y como segunto el valor
    plt.text(i,val,round(val,4),ha="center") #los argumentos de text son X y Y y el valor a a escribir
plt.grid()
plt.xlabel("Conteo de caras")
plt.ylabel("Probabilidad")
plt.show()
```

```python
#probabilidad de 3 caras o menos
accum_freqs = [sum(freqs[:i+1]) for i in range(len(freqs))]
accum_freqs

plt.bar(results,accum_freqs)
for i, val in enumerate(accum_freqs): #enumerate nos devuelve como primer termino el indice de la coleccion, y como segunto el valor
    plt.text(i,val,round(val,4),ha="center") #los argumentos de text son X y Y y el valor a a escribir
plt.grid()
plt.xlabel("Conteo de caras")
plt.ylabel("Probabilidad acumulada")
plt.show()
```

```python
start = 0
end = 10
num = 100

numbers = [start+i*(end-start)/num for i in range(num)] #100 numeros del 1 al 10 igualmente espaciados
#numbers

freqs = [norm.pdf(value,5,1) for value in numbers] # Los argumentos de norm.pdf son numeros de sucesos, la media y desviacion estantar,para valores continuos
#freqs

plt.bar(numbers,freqs)
plt.grid(True)
plt.xlabel("Valores")
plt.ylabel("Probabilidad")
plt.show()
```

```python
#Prueba de hipotesis
sacks = np.array([2.3,2.9,3.1,2.5,2.8,3.0,2.7])
population_mean = 3
population_std = 0.5
sample_mean  = sacks.mean()
samples_std = sacks.std()
print(f'media:{sample_mean:.2f} y desviacion: {samples_std: .2f}')
```

```python
#Prueba de medias
z_score = (sample_mean-population_mean)/(population_std/np.sqrt(len(sacks)))
#numerador diferencia de la media poblacional y muestral
#denominador desviacion de la poblacion entre la raiz cuadrada de la cantidad de muestras(error estandar)
z_score 
```
