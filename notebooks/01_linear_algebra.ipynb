```python
xvalues = np.linspace(-4,4) #crea puntos equidistantes en un intervalo determinado
```

```python
def square(x):
    return x*x
```

```python
square(2)
```

```python
f_xvalues = square(xvalues)
f_xvalues
```

```python
plt.plot(xvalues,f_xvalues)
plt.show
```

```python
def sinfuntion(x):
    return np.sin(x)
```

```python
g_values = sinfuntion(xvalues)
plt.plot(xvalues,g_values)
plt.show
```

```python
def two_the_power(x):
    return np.power(2,x)
```

```python
h_values = two_the_power(xvalues)
plt.plot(xvalues,h_values)
plt.show
```
